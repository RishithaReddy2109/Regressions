```python
x=data['Head Size(cm^3)']  #independent feature
```

```python
y=data['Brain Weight(grams)'] #dependent feature
```

```python
xmean=x.mean()
print(xmean)
```

```python
ymean=y.mean()
print(ymean)
```

```python
nume=0
denom=0
for i in x:
    nume=(x-xmean)*(y-ymean)+nume
    denom=(x-xmean)**2+denom
m=nume/denom
print(m)
```

```python
c=ymean-m*xmean
print(c)
```

```python
arr=[]
for i in range(len(x)):
    arr.append(m*xmean+c)
print(arr)
    
    
```

```python
ypsum=0
n_sum=0
for i in range(len(arr)):
    ypsum=ypsum+(arr[i]-ymean)**2
for i in range(len(y)):
    n_sum=n_sum+(y[i]-ymean)**2
r2=ypsum/n_sum  ## Error rate
print(ypsum)
print(n_sum)
print("Error rate:",r2)
print("Accuracy rate:",1-r2) # accuracy rate
```
